```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
temp = pd.read_csv("DataSets/GlobalLandTemperatures/GlobalTemperatures.csv")
```

```python
temp['dt'] = pd.to_datetime(temp['dt'])
```

```python
fig, ax = plt.subplots(figsize=(16,6 ))
ax.plot(temp['dt'], temp['LandMaxTemperature'], label='Max T', color='red')
ax.plot(temp['dt'], temp['LandMinTemperature'], label='Min T', color='blue')
ax.set_xlabel('Year')
ax.set_ylabel('Temperature (C)')
ax.set_title('Land Max and Min Temperatures')
ax.set_ylim(-10, 30)
plt.show()
```

```python
temp = temp[temp['dt'].dt.year >= 2000]
temp = temp.reset_index(drop=True)
temp.drop(['LandAverageTemperatureUncertainty', 
          'LandMaxTemperatureUncertainty', 
          'LandMinTemperatureUncertainty', 
          'LandAndOceanAverageTemperature',
          'LandAndOceanAverageTemperatureUncertainty'],
          axis=1, inplace=True
         )
temp
```

```python
fig, ax = plt.subplots()
ax.plot(temp['dt'], temp['LandMaxTemperature'], label='Max T', color='red')
ax.plot(temp['dt'], temp['LandMinTemperature'], label='Min T', color='blue')
ax.set_xlabel('Year')
ax.set_ylabel('Temperature (C)')
ax.set_title('Land Max and Min Temperatures')
plt.show()
```

```python
cameroon = pd.read_csv("DataSets/GlobalLandTemperatures/GlobalLandTemperaturesByCountry.csv")
```

```python
cameroon = cameroon[cameroon['Country'] == 'Cameroon']
```

```python
cameroon['dt'] = pd.to_datetime(cameroon['dt'])
cameroon = cameroon[cameroon['dt'].dt.year >= 2000]
cameroon = cameroon.reset_index(drop=True)
cameroon.drop(
              ['AverageTemperatureUncertainty', 
              'Country'], 
              axis=1, inplace=True
             )
cameroon
```

```python
fig, ax = plt.subplots()
ax.plot(cameroon['dt'], cameroon['AverageTemperature'], label='Temperature', color='red')
ax.set_xlabel('Year')
ax.set_ylabel('Temperature (C)')
ax.set_title('Cameroon Average Temperature')
ax.set_ylim(15, 35)
plt.show()
```

```python
cocoa = pd.read_csv("DataSets/Commodities/Cocoa_data.csv")
cocoa
```

```python
cocoa['date'] = pd.to_datetime(cocoa['date'])
cocoa = cocoa[cocoa['date'].dt.year <= 2013]
cocoa = cocoa.reset_index(drop=True)
cocoa.drop(
              ['volume', 
              'open',
              'close'], 
              axis=1, inplace=True
             )
cocoa
```

```python
fig, ax = plt.subplots()
ax.plot(cocoa['date'], cocoa['high'], label='High Price', color='green')
ax.plot(cocoa['date'], cocoa['low'], label='Low Price', color='blue')
ax.set_xlabel('Year')
ax.set_ylabel('Price')
ax.set_title('Cocoa Price')
plt.show()
```

```python
cameroon_sample = cameroon[cameroon['dt'].dt.year == 2007]
cocoa_sample = cocoa[cocoa['date'].dt.year == 2007]
```

```python
fig, ax1 = plt.subplots(figsize=(16,6))
ax1.plot(cocoa_sample['date'], cocoa_sample['high'], label='High Price', color='green')
ax1.plot(cocoa_sample['date'], cocoa_sample['low'], label='Low Price', color='blue')
ax1.set_xlabel('Year')
ax1.set_ylabel('Price')
ax1.set_ylim(500, 4000)

ax2 = ax1.twinx()
ax2.plot(cameroon_sample['dt'], cameroon_sample['AverageTemperature'], label='Temperature', color='red')
ax2.set_ylabel('Temperature (C)', color='red')
ax2.set_ylim(15, 35)

ax1.set_title('Price and Temperature over Year')
plt.show()
```

```python
cotton = pd.read_csv("DataSets/Commodities/Cocoa_data.csv")
cotton
```

```python
cotton['date'] = pd.to_datetime(cotton['date'])
```

```python
cotton = cotton[cotton['volume'] != 0]
cotton
```

```python
fig, ax1 = plt.subplots(figsize=(16,6))
ax1.plot(cotton['date'], cotton['high'], label='High Price', color='lightblue')
ax1.plot(cotton['date'], cotton['low'], label='Low Price', color='blue')
ax1.set_xlabel('Year')
ax1.set_ylabel('Price')
ax1.set_ylim(500, 4000)

ax2 = ax1.twinx()
ax2.plot(cotton['date'], cotton['volume'], label='Volume', color='orange')
ax2.set_ylabel('Volume')
ax2.set_ylim(-1000, 100000)

ax1.set_title('Price and Trade Volume')
plt.show()
```

```python
plt.style.use('_mpl-gallery')

x = cotton['date']
y = np.vstack([cotton['high'], cotton['low'], cotton['volume']])

fig, ax = plt.subplots(figsize=(16, 5))

ax.stackplot(x, y)

ax.set(ylim=(0, 40000))
ax.set_xlim(x.min(), x.max())
```
